## Data Visualization

```python
countries=df.country_txt.value_counts().head(10)
plt.figure(figsize=(17,5))
plt.bar(countries.index,countries.values,color='red')
plt.title('Top 10 Countries with most terrorism')
plt.xlabel('Country')
plt.ylabel('Counts')
plt.show()
plt.style.use('seaborn')
```

```python
countries=df.country_txt.value_counts().tail(10)
topcountries=list(countries)
plt.figure(figsize=(16,5))
plt.bar(countries.index,countries.values,color='green')
plt.title('Top 10 Countries with Least terrorism')
plt.xlabel('Country')
plt.ylabel('Counts')
plt.show()
plt.style.use("ggplot")
```

## Top 10 Countries Terrorism Occur Ratio Wise

```python
country=df.country_txt.value_counts().head(10)
plt.figure(figsize=(16,5))
plt.pie(country.values,labels=country.index,radius=2,autopct='%0.1f%%')
plt.show()
plt.style.use("seaborn")
```

## Top Most Terrorism Attack Years

```python
topyears=df.groupby('iyear')['eventid'].agg(['count']).reset_index().sort_values(by='count',ascending=False).head(10)
print(topyears)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(df["iyear"],order=df['iyear'].value_counts(ascending=False).head(10).index,palette="dark")
plt.xlabel("Year",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Top Most Terrorism Attack Years",fontsize=20)
plt.show()
```

## Top Few Terrorism Attack Years

```python
lessyears=df.groupby('iyear')['eventid'].agg(['count']).reset_index().sort_values(by='count',ascending=True).head(10)
print(lessyears)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(df["iyear"],order=df['iyear'].value_counts(ascending=False).tail(10).index,palette="bright")
plt.xlabel("Year",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Top Less Terrorism Attack Years",fontsize=20)
plt.show()
```
